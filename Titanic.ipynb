```
import pandas as pd

df_train = pd.read_csv('train.csv')
df_test = pd.read_csv('test.csv')

display(df_train.head())
display(df_test.head())
```

```
print("Shape of df_train:", df_train.shape)
print("\nInfo of df_train:")
df_train.info()
print("\nDescribe df_train:")
display(df_train.describe())
print("\nSurvived value counts:")
print(df_train['Survived'].value_counts())


print("\nShape of df_test:", df_test.shape)
print("\nInfo of df_test:")
df_test.info()
print("\nDescribe df_test:")
display(df_test.describe())


def missing_percentage(df):
    missing_values = df.isnull().sum()
    missing_percentage = (missing_values / len(df)) * 100
    return missing_percentage

print("\nMissing values percentage in df_train:")
print(missing_percentage(df_train))

print("\nMissing values percentage in df_test:")
print(missing_percentage(df_test))
```

```
df_train['Age'].fillna(df_train['Age'].median(), inplace=True)
df_test['Age'].fillna(df_test['Age'].median(), inplace=True)
df_train['Embarked'].fillna(df_train['Embarked'].mode()[0], inplace=True)
df_test['Fare'].fillna(df_test['Fare'].median(), inplace=True)

df_train['Has_Cabin'] = df_train['Cabin'].notna().astype(int)
df_train.drop('Cabin', axis=1, inplace=True)
df_test['Has_Cabin'] = df_test['Cabin'].notna().astype(int)
df_test.drop('Cabin', axis=1, inplace=True)


df_train.drop_duplicates(inplace=True)
df_test.drop_duplicates(inplace=True)


def missing_percentage(df):
    missing_values = df.isnull().sum()
    missing_percentage = (missing_values / len(df)) * 100
    return missing_percentage

print("Missing values percentage in df_train:")
print(missing_percentage(df_train))
print("\nMissing values percentage in df_test:")
print(missing_percentage(df_test))
```

```
import matplotlib.pyplot as plt
import seaborn as sns

survival_by_class = df_train.groupby('Pclass')['Survived'].mean()
print("Survival rate by passenger class:\n", survival_by_class)

survival_by_sex = df_train.groupby('Sex')['Survived'].mean()
print("\nSurvival rate by gender:\n", survival_by_sex)

df_train['AgeGroup'] = pd.cut(df_train['Age'], bins=[0, 18, 60, 100], labels=['Child', 'Adult', 'Senior'])
survival_by_age = df_train.groupby('AgeGroup')['Survived'].mean()
print("\nSurvival rate by age group:\n", survival_by_age)

survival_by_sibsp = df_train.groupby('SibSp')['Survived'].mean()
print("\nSurvival rate by SibSp:\n", survival_by_sibsp)

survival_by_parch = df_train.groupby('Parch')['Survived'].mean()
print("\nSurvival rate by Parch:\n", survival_by_parch)


df_train['FareGroup'] = pd.qcut(df_train['Fare'], 4, labels=['Low', 'Medium', 'High', 'Very High'])
survival_by_fare = df_train.groupby('FareGroup')['Survived'].mean()
print("\nSurvival rate by Fare:\n", survival_by_fare)

survival_by_embarked = df_train.groupby('Embarked')['Survived'].mean()
print("\nSurvival rate by Embarked:\n", survival_by_embarked)

survival_by_class_sex = df_train.groupby(['Pclass', 'Sex'])['Survived'].mean().unstack()
print("\nSurvival rate by Passenger class and Sex:\n", survival_by_class_sex)

plt.figure(figsize=(10, 6))
sns.barplot(x='Pclass', y='Survived', data=df_train)
plt.title('Survival Rate by Passenger Class')
plt.show()

plt.figure(figsize=(10,6))
sns.barplot(x='Sex', y='Survived', data=df_train)
plt.title('Survival Rate by Sex')
plt.show()
```

```
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(15, 10))
for i, col in enumerate(['Age', 'Fare', 'SibSp', 'Parch']):
    plt.subplot(2, 2, i + 1)
    sns.displot(data=df_train, x=col, hue='Survived', kind='hist', kde=True)
    plt.title(f'Distribution of {col}')
plt.tight_layout()
plt.show()

plt.figure(figsize=(15, 10))
for i, col in enumerate(['Age', 'Fare', 'SibSp', 'Parch']):
    plt.subplot(2, 2, i + 1)
    sns.boxplot(x='Survived', y=col, data=df_train)
    plt.title(f'Box plot of {col} by Survival')
plt.tight_layout()
plt.show()

categorical_cols = ['Pclass', 'Sex', 'Embarked', 'AgeGroup', 'FareGroup']
plt.figure(figsize=(15, 15))
for i, col in enumerate(categorical_cols):
    plt.subplot(3, 2, i + 1)
    sns.countplot(x=col, hue='Survived', data=df_train)
    plt.title(f'Survival Rate by {col}')
    plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

plt.figure(figsize=(10, 8))
sns.heatmap(df_train[['Survived', 'Age', 'Fare', 'SibSp', 'Parch']].corr(), annot=True, cmap='coolwarm')
plt.title('Correlation Matrix of Numerical Features')
plt.show()
```
