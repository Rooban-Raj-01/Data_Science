```
import csv

try:
  data = []
  with open('API_SP.POP.TOTL_DS2_en_csv_v2_76253.csv', 'r', encoding='utf-8') as file:
    reader = csv.reader(file)
    for row in reader:
      data.append(row)
  df_main = pd.DataFrame(data)
  print("Successfully loaded using the csv module.")
  display(df_main.head())
except Exception as e:
  print(f"Failed to load using csv module: {e}")
  print("Cannot proceed without a correctly loaded DataFrame.")
```

```
header = df_main.iloc[4].values

df_main = pd.DataFrame(df_main.values[5:], columns=header)

display(df_main.head())
```

```
df_visualization = df_main[['Country Name', '2022']].copy()

df_visualization['2022'] = pd.to_numeric(df_visualization['2022'], errors='coerce')

median_2022 = df_visualization['2022'].median()
df_visualization['2022'] = df_visualization['2022'].fillna(median_2022)

df_visualization = df_visualization.rename(columns={'2022': 'Population'})

display(df_visualization.head())
```

```
import matplotlib.pyplot as plt

# Create the histogram
plt.figure(figsize=(12, 6))  # Adjust figure size for better readability
plt.hist(df_visualization['Population'], bins=50, color='skyblue', edgecolor='black')
plt.yscale('log') # Use logarithmic scale for y-axis
plt.title('Distribution of Population Totals in 2022', fontsize=16)
plt.xlabel('Population', fontsize=14)
plt.ylabel('Number of Countries', fontsize=14)
plt.grid(axis='y', alpha=0.75)
plt.show()
```
